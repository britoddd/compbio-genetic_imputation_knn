```python
from Bio.Seq import Seq, MutableSeq
```

```python
seq_A = Seq('ATCG')
seq_B = Seq('_T_G')
seq_C = Seq('ATCA')
seq_D = Seq('GTCG')
```

```python
def similarity_score(target, seq):
  total_same = 0
  for allele_target, allele_seq in zip(target, seq):
    if allele_target == '_':
      continue
    if allele_target == allele_seq:
      total_same += 1
  return total_same
```

```python
def calculate_similarity(target, seqs):
  calculated_seqs = {}
  for seq in seqs:
    calculated_seqs[seq] = similarity_score(target, seq)
  return calculated_seqs
```

```python
def KNN_imputation(k, target, seqs):
  calculated_seqs = calculate_similarity(target, seqs)
  sorted_seqs = sorted(calculated_seqs.items(), key=lambda x: x[1], reverse=True)
  knn_seqs = sorted_seqs[:min(k, len(seqs))]

  result = MutableSeq(target)

  print('Allele Numbers Comparison')
  for i, allele in enumerate(result):
    if allele == '_':
      seq_total = {'C': 0, 'G': 0, 'A': 0, 'T': 0}
      for seq, _ in knn_seqs:
        snp = seq[i]
        seq_total[snp] += 1
      knn_snp = max(seq_total, key=seq_total.get)
      print(f'i = {i} -> {seq_total} => {knn_snp}')
      result[i] = knn_snp
  return result
```

```python
K_value = 3
target = seq_B
sequences = [seq_A, seq_C, seq_D]

print('Target          :', target)
print('=' * 50)
knn_result = KNN_imputation(K_value, target, sequences)
print('=' * 50)
print('Sequence Result :', knn_result)
```
